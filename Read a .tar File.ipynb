```python
import torch

# Path to your file
tar_path = r"H:\University\4-1\RME-4100\Project\checkpoints\miniimagenet\WideResNet28_10_S2M2_R\470.tar"

# Load the checkpoint
checkpoint = torch.load(tar_path, map_location='cpu')  # Trusted source = safe to load

# Display top-level keys
print("Top-level keys:", checkpoint.keys())

# View the epoch it was saved at
print("Saved at epoch:", checkpoint.get("epoch"))

# View model state_dict keys (parameters)
print("\nModel state_dict keys:")
for key in checkpoint['state'].keys():
    print("  ", key)

# Optional: Inspect classifier object
print("\nClassifier type:", type(checkpoint.get("classifier")))
```
